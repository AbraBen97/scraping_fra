```python
import pandas as pd
import numpy as np

donnee = pd.read_csv('donnees.csv')


```

```python
def horaire(strin):
    strin = strin.replace("%", "")
    strin = float(strin)
    return strin

horaire = np.vectorize(horaire)
```

```python
donnee.nuit = horaire(donnee.nuit)
donnee['nuit'] = (donnee['nuit'] >= 50).astype(int)


donnee.jour = horaire(donnee.jour)
donnee['jour'] = (donnee['jour'] >= 50).astype(int)


donnee.hiver = horaire(donnee.hiver)
donnee['hiver'] = (donnee['hiver'] >= 50).astype(int)


donnee.printemps	= horaire(donnee.printemps)
donnee['printemps'] = (donnee['printemps'] >= 50).astype(int)


donnee.ete	= horaire(donnee.ete)
donnee['ete'] = (donnee['ete'] >= 50).astype(int)


donnee.autone	= horaire(donnee.autone)
donnee['autone'] = (donnee['autone'] >= 50).astype(int)
```

```python
donnee['women'] = 0
donnee['men'] = 0

for i in range(donnee.shape[0]):
    if "women" in donnee.genre.iloc[i,]:
        donnee.loc[i, "women"] = 1
    else:
        donnee.loc[i, "women"]  = 0
    
    if " men" in donnee.genre.iloc[i,]:
        donnee.loc[i, "men"] = 1
    else:
        donnee.loc[i, "men"] = 0  
```

```python
donnee['composantes'] = donnee['composantes'].astype(str)
donnee.shape[0]
```

```python
donnee = donnee.dropna()
donnee.shape[0]
```

```python
lis = donnee.composantes
liste_element = []
for i in lis:
    try:
        tab = i.split(" - ")
        for j in tab:
          liste_element.append(j)
    except:
        print(i)
        
        
liste_element = set(liste_element)
liste_element.remove("")
        

        
```

```python
L = []
# Créer une copie de la liste pour itérer dessus
copie_liste_element = liste_element.copy()

# Itérer sur la copie de la liste
for i in copie_liste_element:
    if "-" in i:
        # Ajouter l'élément à la liste L
        L.append(i)
        # Supprimer l'élément de la liste d'origine
        liste_element.remove(i)

print(L)
```

```python
B = L[2].split("-")
C = [i.strip() for i in B]
C.remove("")
liste_element.update(C)
len(liste_element)
```

```python
import pandas as pd

# Fonction pour ajouter une ligne avec 0 ou 1
def ajout(liste):
    # Créer un dictionnaire pour stocker les valeurs 0 ou 1
    row_data = {}
    # Séparer les éléments de la chaîne
    
    for j in liste:
        elements = j.split(" - ")
    
        # Remplir le dictionnaire
        for i in liste_element:
         # Mettre 1 si l'élément est présent dans la chaîne, sinon 0
            row_data[i] = 1 if i in elements else 0
    
    # Ajouter la nouvelle ligne au DataFrame
        df1.loc[len(df1)] = row_data

```

## MISE EN PLACE DE LA RECOMMANDATION 

```python
import scipy
from scipy.spatial.distance import pdist, squareform
```

```python
def prediction(*parfums):
    global df_concat
    new_user = 0
    for parfum in parfums:
        new_user = new_user + df_concat.loc[parfum]
    new_user[new_user > 1] = 1
    df_concat.loc['new_user'] = new_user
    
    jaccard_distances = pdist(df_concat.values, metric='jaccard')
    jaccard_similarity_array = 1 -  squareform(jaccard_distances)
    
    jaccard_similarity_df = pd.DataFrame(jaccard_similarity_array, index=df_concat.index, columns=df_concat.index)
    jaccard_similarity_series = jaccard_similarity_df.loc['new_user']

# Sort these values from highest to lowest

    indexes_to_remove = list(parfums)

# Utilisez la méthode .drop() pour supprimer les éléments
    jaccard_similarity_series = jaccard_similarity_series.drop(indexes_to_remove)
    ordered_similarities = jaccard_similarity_series.sort_values(ascending=False)
    
    
    
    df_concat = df_concat.drop(index='new_user', axis=0)
    L = ordered_similarities[1:6]
    dic = { L.index[0]: L.values[0],L.index[1]: L.values[1],L.index[2]: L.values[2],L.index[3]: L.values[3],
           L.index[4]: L.values[4]
        
    }
    
    return dic
    
    
    
    
   
```

```python
prediction('Soleil Vibrant Lalique for women',"Noble Rose 4711 for women and men","Ubud Retreat Yu Parfums for women and men")
```
