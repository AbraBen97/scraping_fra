```python
# Creaton de data frame resumant les match pour chacune des equipe

Maroc = (resume [(resume['Team_1'] == 'Maroc') | (resume['Team_2'] == 'Maroc')])[["Team_1","Team_2","Goal_Team_1","Goal_Team_2","Tirs_Team_1"]]
Tanzanie = (resume [(resume['Team_1'] == 'Tanzanie') | (resume['Team_2'] == 'Tanzanie')])[["Team_1","Team_2","Goal_Team_1","Goal_Team_2","Tirs_Team_1"]]
Zambie =(resume [(resume['Team_1'] == 'Zambie') | (resume['Team_2'] == 'Zambie')])[["Team_1","Team_2","Goal_Team_1","Goal_Team_2","Tirs_Team_1"]]
RD_Congo=(resume [(resume['Team_1'] == 'RD Congo') | (resume['Team_2'] == 'RD Congo')])[["Team_1","Team_2","Goal_Team_1","Goal_Team_2","Tirs_Team_1"]]
```

```python
#definition d'une fonction pour ajouter la colone victoire, donnant le vainqueur 
# de chaque match

def resultats(df):

    for i in range(len(df)):
        if int(df.iloc[i,2]) > int(df.iloc[i,3]):
            df.iloc[i,4] = df.iloc[i,0]
        elif int(df.iloc[i,2]) < int(df.iloc[i,3]):
            df.iloc[i,4] = df.iloc[i,1]
        else :
            df.iloc[i,4] = "nul"
        df.rename(columns={"Tirs_Team_1":"victoire"},inplace=True) 
    return df
     
    
```

```python
Tanzanie = resultats(Tanzanie)
Maroc = resultats(Maroc)
Zambie = resultats(Zambie)
RD_Congo = resultats(RD_Congo)
Tanzanie
```

```python
#fonction pour determiner le nombre de but marque ert conceder pour une equipe

def nbr_but(dt, pays):
    but_maqu = 0
    but_encai = 0
    for i in range(len(dt)):
        if (dt.iloc[i,0])== pays: 
            a = dt.iloc[i,2]
            b=dt.iloc[i,3]
        else:
            b = dt.iloc[i,2]
            a= dt.iloc[i,3]
        but_maqu = but_maqu + int(a)
        but_encai = but_encai + int(b)
    return [but_maqu, but_encai]
```

```python
#fonction qui donne le resume pour une equipe

def resume_equipe(df, pays):
    but = nbr_but(df, pays)
    gagne = len(df[df["victoire"]==pays])
    nul = len(df[df["victoire"]=="nul"])
    defaite = abs(gagne-nul)
    dic = {"equipe":pays,"MJ":len(df), 'G':gagne,"N":nul, "P": defaite,  
                "BP":but[0], "BC":but[1] , "DB":but[0]-but[1],
                "Pts": 3*gagne+ nul
                }
    return dic
```

```python
#Tableau recapitulatif de la poule 

columns = [
    "equipe",'MJ', 'G', 'N', 'P', 'BP',
    'BC', 'DB', 'Pts'] 

tableau = pd.DataFrame(columns=columns)

d1 = resume_equipe(Maroc, "Maroc")
d2 = resume_equipe(Tanzanie, "Tanzanie")
d3 = resume_equipe(Zambie, "Zambie")
d4 = resume_equipe(RD_Congo, "RD Congo")
for i in [d1,d2,d3,d4]:
    tableau.loc[len(tableau)] = i

```

```python
tableau = tableau.set_index('equipe')
tableau = tableau.sort_values(by=['Pts','DB'], ascending=False)

tableau
```
